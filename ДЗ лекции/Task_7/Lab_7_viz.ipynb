```
fig = go.Figure([go.Histogram(x=df['sepal_length'])])
fig.show()
```

```
fig = go.Figure()
fig.add_trace(go.Box(x=df['sepal_length'],name='sepal_length'))
fig.add_trace(go.Box(x=df['sepal_width'],name='sepal_width'))
fig.show()
```

```
df.groupby('species').mean().sepal_width
```

```
trace1 = go.Bar(
                x = df.species.unique(),
                y = df.groupby('species').mean().sepal_width,
                name = "sepal_width",
                marker = dict(color = 'rgb(102,255,255)'),
                text = df.groupby('species').mean()['sepal_width'])
trace2 = go.Bar(
                x = df.species.unique(),
                y = df.groupby('species').mean().sepal_length,
                name = "sepal_length",
                marker = dict(color = 'rgb(102,178,255)'),
                text = df.groupby('species').mean()['sepal_length'])
trace3 = go.Bar(
                x = df.species.unique(),
                y = df.groupby('species').mean().petal_length,
                name = "petal_length",
                marker = dict(color = 'rgb(102,102,255)'),
                text = df.groupby('species').mean()['petal_length'])
trace4 = go.Bar(
                x = df.species.unique(),
                y = df.groupby('species').mean().petal_width,
                name = "petal_width",
                marker = dict(color = 'rgb(178, 102, 255)'),
                text = df.groupby('species').mean()['petal_width'])

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(title = 'Line Plot: Mean Values by Species',
              xaxis= dict(title= 'Species',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Mean Values',ticklen= 5,zeroline= False)
             )
fig = go.Figure(data = data, layout = layout)
url = py.plot(fig, validate=False)
fig.show()
```
