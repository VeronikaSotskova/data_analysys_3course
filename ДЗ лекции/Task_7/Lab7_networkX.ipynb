```
import networkx as nx

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

## Задание  1
Создайте граф и добавьте в него $5$ вершин с городами России, у каждой должен быть атрибут (характеристика) `time_diff`, означающий временную разницу с Москвой в часах.

```
# Создаём заготовку графа, в которую будут добавляться вершины и рёбра.
g = nx.Graph()
```

```
g.add_node('Kazan', time_diff=0)
g.add_node('Samara', time_diff=1)
g.add_node('Spb', time_diff=0)
g.add_node('Omsk', time_diff=3)
g.add_node('Moscow', time_diff=0)
```

## Задание 2
Выберите какую-нибудь конкретную авиакомпанию (Аэрофлот, S7, Utair, etc).

В созданный граф добавьте рёбра между выбранными $5$ю городами, между которыми есть авиарейсы данной компании. У каждого ребра должны быть следующие атрибуты:
- расстояние между городами в километрах `distance`
- время перелёта в минутах `time`

При желании можно добавить в граф также несколько городов, являющихся пересадочными пунктами.

```
g.add_edge('Kazan', 'Samara', distance=296, time=270)
g.add_edge('Kazan', 'Moscow', distance=717, time=105)
g.add_edge('Kazan', 'Omsk', distance=1527, time=140)
g.add_edge('Moscow', 'Samara', distance=853, time=110)
g.add_edge('Moscow', 'Omsk', distance=2235, time=240)
g.add_edge('Spb', 'Moscow', distance=634, time=95)
g.add_edge('Spb', 'Kazan', distance=1198, time=125)
```

```
pos=nx.spring_layout(g)
nx.draw_networkx(g,pos,node_color='lightblue')
labels = nx.get_edge_attributes(g,'distance')
nx.draw_networkx_edge_labels(g,pos,edge_labels=labels)
plt.axis('off')
```

```
pos=nx.spring_layout(g)
nx.draw_networkx(g,pos,node_color='lightblue')
labels = nx.get_edge_attributes(g,'time')
nx.draw_networkx_edge_labels(g,pos,edge_labels=labels)
plt.axis('off')
```

## Задание 3 
Выберите любую пару городов и постройте между ними следующие кратчайшие пути:
- без взвешивания
- со взвешиванием по расстоянию
- со взвешиванием по времени

```
# Без учёта весов.
print(nx.shortest_path(g, 'Spb', 'Samara'))
```

```
# С учётом весов: distance.
print(nx.shortest_path(g, 'Spb', 'Samara', weight='distance'))
```

```
# С учётом весов: distance.
print(nx.shortest_path(g, 'Spb', 'Samara', weight='time'))
```
